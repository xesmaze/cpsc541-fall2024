```python
import matplotlib.pyplot as plt
import numpy as np

# Define the seed sizes and their respective frequencies (standard deviation from mean)
seed_groups = ['-2σ', '-1σ', '0σ', '+1σ', '+2σ']
frequencies = [5, 10, 15, 10, 5]

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create scatter plots for each seed group
for i, (group, freq) in enumerate(zip(seed_groups, frequencies)):
    x = np.random.normal(i, 0.1, size=freq)
    y = np.random.normal(0, 0.1, size=freq)
    ax.scatter(x, y, label=f"{group}", alpha=0.6)

# Add a simple normal distribution curve to represent the statistical scale
x_curve = np.linspace(-2, 2, 100)
y_curve = np.exp(-0.5 * x_curve**2) / np.sqrt(2 * np.pi)
#ax.plot(x_curve, y_curve, color='black', lw=2, label="Normal Distribution")

# Customizing the plot
ax.axhline(0, color='grey', lw=0.5)
ax.axvline(0, color='grey', lw=0.5)
ax.set_xticks(range(len(seed_groups)))
ax.set_xticklabels(seed_groups)
ax.set_yticks([])
ax.set_xlim(-0.5, len(seed_groups)-0.5)
ax.set_ylim(-0.5, 0.5)
ax.set_xlabel("Seed Size (Standard Deviations)")
ax.set_ylabel("Frequency")
ax.set_title("Visualization Inspired by Galton's Seed Distribution Study")
ax.legend(title="Seed Size Groups")

# Show plot
plt.show()
```
