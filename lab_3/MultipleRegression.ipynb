```python
! pip install numpy matplotlib scikit-learn

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Simulate data
np.random.seed(42)  # For reproducibility
x1 = np.random.rand(100) * 10  # Independent variable 1
x2 = np.random.rand(100) * 10  # Independent variable 2
y = 3.5 * x1 + 2.5 * x2 + 5 + np.random.randn(100)  # Dependent variable with some noise

# Prepare the data for fitting
X = np.column_stack((x1, x2))

# Perform multiple linear regression
model = LinearRegression()
model.fit(X, y)

# Coefficients and intercept
print(f"Intercept (b): {model.intercept_}")
print(f"Coefficients (a1, a2): {model.coef_}")

# Generate predictions for the fitted plane
y_pred = model.predict(X)

# Visualization
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(x1, x2, y, color='blue', label='Data Points')

# Plot the regression plane
x1_grid, x2_grid = np.meshgrid(np.linspace(x1.min(), x1.max(), 50),
                               np.linspace(x2.min(), x2.max(), 50))
y_grid = model.intercept_ + model.coef_[0] * x1_grid + model.coef_[1] * x2_grid
ax.plot_surface(x1_grid, x2_grid, y_grid, color='red', alpha=0.5)

# Labels
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.title('Multiple Linear Regression Visualization')
plt.legend()
plt.show()
```
